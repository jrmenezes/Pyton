**Dados abertos da Camera do deputados**

https://dadosabertos.camara.leg.br/swagger/api.html

```python
import requests
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
url        = 'https://dadosabertos.camara.leg.br/api/v2/deputados/'
resposta   = requests.request("GET", url)
objetos    = json.loads(resposta.text)
dados      = objetos['dados']
#dados_parametros = objetos['dados']
```

```python
df = pd.DataFrame(dados)
```

***A partir dos dados acima será feita uma análise para identificar o conjuto de atributos que servirá como padrão para classificão do dados.***

```python
# Análise da quantidade de deputados pela UF
 
x = df['siglaUf'].value_counts().index
y = df['siglaUf'].value_counts().values
plt.figure(figsize = (15, 5))
sns.barplot(x=x, y=y, palette="deep")
plt.plot()
```
